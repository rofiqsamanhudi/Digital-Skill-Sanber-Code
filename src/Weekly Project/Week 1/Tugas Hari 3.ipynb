<h2>Soal 2: Built-in Function</h2>

Jawab Pertanyaan di bawah ini:
    
Apa itu built-in Function?

Sebutkan 3 built-in Function di pyhon!

Berikan contoh masing2 1 cara penggunaan built-in function yang kamu sebutkan sebelumnya

<!-- Apa itu built-in Function? -->
Built-in function adalah fungsi yang sudah disediakan oleh Python dan dapat digunakan langsung tanpa perlu mendefinisikannya terlebih dahulu.

<!-- Sebutkan 3 built-in Function di python! -->
sum()
print()
type()
len()

```python
# Contoh penggunaan built in function
# sum()
angka = [1, 2, 3, 4, 5]
jumlah = sum(angka)

# print()
print("Jumlah:", jumlah)

# type()
print(type(angka))
print(type(jumlah))

# len()
panjang_string =len(angka)
print("panjang_string:", panjang_string)
```

<h2>Soal 5: Membuat Simple Function</h2>

Buatlah suatu fungsi yang menerima satu input argumen berbentuk list dan mempunyai elemen bertipe numeric semua, dimana fungsi tersebut berguna untuk menghitung rata2 dari kumpulan elemen list tersebut. namai fungsi tersebut 'mean_list'

```python
obj_list = [11.25, 18.0, 20.0, 10.75, 9.50, 13.45, 23.0, 7.0, 8.45, 2.50]
def mean_list(inp_list):
 total = sum(inp_list)
 mean = total/len(inp_list)
 return round(mean, 2)

print(mean_list(obj_list))
```

Expected Output:

12.39

<h2>Soal 6: Membuat Function dengan Multiple arguments</h2>

Buatlah suatu fungsi untuk melakukan penggabungan antara dua list 

```python
obj_list = [1, 2, 3, 4]
obj_penambah = [5, 6, 7]

def kali_list(inp_list1, inp_list2):
 combine = inp_list1 + inp_list2
 return combine

print(kali_list(obj_list, obj_penambah))
```

# Soal 7. Mapping Fungsi lambda

Hitunglah jumlah bilangan genap dari sebuah objek list berikut menggunakan lambda function

```python
obj_list = [i for i in range(1000)]
hasil = sum(1 for x in obj_list if x % 2 == 0)

print(hasil)
```

Expected output:

500
