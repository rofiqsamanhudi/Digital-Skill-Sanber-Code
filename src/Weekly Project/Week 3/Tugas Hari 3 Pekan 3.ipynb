<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(2*-np.pi, 2*np.pi, 200)

tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# Plotting
fig, ax = plt.subplots(1, 3, figsize=(18, 7))

ax[0].plot(x, tan, 'b-', label='tanX')
ax[0].set_xticks([])
ax[0].set_yticks([])

ax[1].plot(x, sin, 'r-', label='sinX')
ax[1].set_xticks([])
ax[1].set_yticks([])

ax[2].plot(x, cos, 'g-', label='cosX')
ax[2].set_xticks([])
ax[2].set_yticks([])

fig.legend(labels=['tanX', 'sinX', 'cosX'], loc='upper center', ncol=1)
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
#code here
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(x, y)
ax.set_xticks([])
ax.set_yticks([])

ins1 = plt.axes([0.075, 0.8, 0.2, 0.2])
ins1.plot(x, y, 'b-', label='CosX')
ins1.plot(x, y5, 'y-', label='CosX^5')
ins1.set_title('Grafik CosX^5')

ins2 = plt.axes([0.775, 0.8, 0.2, 0.2])
ins2.plot(x, y, 'b-', label='CosX')
ins2.plot(x, y4, 'k-', label='CosX^4')
ins2.set_title('Grafik CosX^4')

ins3 = plt.axes([0.075, 0.05, 0.2, 0.2])
ins3.plot(x, y, 'b-', label='CosX')
ins3.plot(x, y3, 'r-', label='CosX^3')
ins3.set_title('Grafik CosX^2')

ins4 = plt.axes([0.775, 0.05, 0.2, 0.2])
ins4.plot(x, y, 'b-', label='CosX')
ins4.plot(x, y2, 'g-', label='CosX^2')
ins4.set_title('Grafik CosX^3')

plt.show()

```
