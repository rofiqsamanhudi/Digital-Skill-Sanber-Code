<h2>Soal 1: Pemahaman Component Matplotlib</h2>

- Jelaskan apa itu Figure
    - Figure adalah wadah atau kanvas utama di mana semua elemen plot berada. Berisi satu atau lebih subplot, serta elemen seperti judul, legend, dan anotasi lainnya.
- Jelaskan apa itu Axis
    - Axis adalah area individual atau subbagian dari figure di mana data digambarkan. Axis mencakup elemen seperti sumbu x dan y, label sumbu, ticks, dan plot data itu sendiri. Dalam satu Figure, bisa terdapat satu atau lebih Axis.

<h2>Soal 2: Membuat Component Figure dan Axis</h2>

- Buatlah Figure dan Axis kosong tanpa Data
- Plot data di bawah ini dengan 2 cara, yaitu dengan membuat figure dan axis secara explicit dan Implicit

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 20, 100)
y = np.cos(x/2)
```

```python
# membuat Figure dan Axis dengan Data Kosong
fig, ax = plt.subplots()
```

```python
#eksplisit
fig, ax = plt.subplots()
ax.plot(x, y, label='cos(x/2)')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Pendekatan Eksplisit')
plt.legend()
plt.show()
```

```python
#implicit
fig, ax = plt.subplots()
plt.plot(x, y, label='cos(x/2)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Pendekatan Implisit')
plt.legend()
plt.show()
```

<h2>Soal 3: Memasukan multiple data ke dalam satu axis</h2>

- Plot ketiga data berikut ke dalam satu axis, dan gunakan seaborn style
- save data tersebut dengan nama file bebas (apa saja) dan tampilkan kembali gambar tersebut.

```python
x = np.linspace(0, 20, 100)
y = np.cos(x/2)
y2 = np.sin(x)
y3 = np.sin(2*x)
```

```python
plt.plot(x, np.sin(y), color='blue')
plt.plot(x, np.sin(y2), color='green')
plt.plot(x, np.sin(y3), color='red')
plt.show()
```
