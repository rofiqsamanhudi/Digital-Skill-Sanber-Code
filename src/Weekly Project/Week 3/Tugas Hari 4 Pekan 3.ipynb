<h2>Soal 2: Perbandingan Kuantitatif Pieplot - Part a whole using Pie plot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menghasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

- Insight


```python
#code here
df_groupby = df.groupby('Genre')['Global_Sales'].mean()
plt.pie(df_groupby, labels=df_groupby.index, autopct='%1.1f%%')
plt.show()
```

<h2>Soal 3: Preparing Data</h2>

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk setiap Region Kecuali Global_Sales berdasarkan kategori Genre.

```python
#code here
genre_groups = df.groupby('Genre')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].mean()
genre_groups
```

<h2>Soal 4: Perbandingan Kuantitatif Barplot - Grouping Visualisasi Dengan Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menghasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
# Code here
average_region = genre_groups.transpose()

fig, ax = plt.subplots(figsize=(16, 8))

ax.set_facecolor('lightgray')

average_region.plot(kind='bar', ax=ax)

plt.title('Mean Sales Video Games By Genre')
plt.xlabel('Region Sales')
plt.ylabel('Mean Sales')
plt.legend(title='Genre')
plt.xticks(rotation=0)

plt.grid(color='white')

plt.show()
```
