<h2>Soal 1: Penggabungan Data</h2>

- Sebutkan perbedaan antara full outer join, inner join, dan left join

1. Full Outer Join
- Definisi: Menggabungkan semua baris dari kedua tabel, baik yang memiliki kecocokan (match) maupun yang tidak.
- Hasil: Menghasilkan baris yang cocok dari kedua tabel serta baris yang tidak cocok dari kedua tabel. Baris yang tidak memiliki pasangan di tabel lain akan memiliki nilai NULL (atau NaN dalam Pandas) di kolom yang tidak ada pasangannya.
- Penggunaan: Berguna ketika ingin melihat semua data dari kedua tabel dan melihat di mana data tidak cocok.

2. Inner Join
- Definisi: Menggabungkan hanya baris yang memiliki kecocokan (match) di kedua tabel.
- Hasil: Menghasilkan hanya baris yang memiliki pasangan di kedua tabel. Baris yang tidak cocok di salah satu tabel tidak akan disertakan dalam hasil.
- Penggunaan: Berguna ketika hanya memerlukan baris yang memiliki data yang cocok di kedua tabel.

3. Left Join (Left Outer Join)
- Definisi: Menggabungkan semua baris dari tabel kiri (left table) dan hanya baris yang cocok dari tabel kanan (right table). Jika tidak ada - kecocokan di tabel kanan, nilai pada kolom dari tabel kanan akan menjadi NULL (atau NaN dalam Pandas).
- Hasil: Menghasilkan semua baris dari tabel kiri dan baris yang cocok dari tabel kanan. Baris di tabel kiri yang tidak memiliki pasangan di tabel kanan akan menampilkan NULL (atau NaN) pada kolom tabel kanan.
- Penggunaan: Berguna ketika ingin melihat semua data dari tabel kiri, dan hanya data yang cocok dari tabel kanan.

<h2>Soal 3: Penggabungan Data</h2>

Lakukan jenis penggabungan data full outer join, inner join, left outer join dan right outer join pada kedua data berikut: 

```python
import pandas as pd

data1 = {
    'id' : [1, 2, 3, 4],
    'Feature1' : ['a1', 'a2', 'a3', 'a4'],
    'Feature2' : ['b1', 'b2', 'b3', 'b4']
}

data2 = {
    'id' : [3, 4, 5 ,6, 7],
    'Feature3' : ['c1', 'c2', 'c3', 'c4', 'c5']
}

df1= pd.DataFrame(data1)
df2= pd.DataFrame(data2)

print(df1)
print(df2)
```

```python
#join data
df_outer_join = pd.merge(df1, df2, on='id', how='outer')
print("\nFull Outer Join:")
print(df_outer_join)

df_inner_join = pd.merge(df1, df2, on='id', how='inner')
print("\nInner Join:")
print(df_inner_join)

df_left_join = pd.merge(df1, df2, on='id', how='left')
print("\nLeft Outer Join:")
print(df_left_join)

df_right_join = pd.merge(df1, df2, on='id', how='right')
print("\nRight Outer Join:")
print(df_right_join)
```

<h2>Soal 5: Process Grouping</h2>

Jelaskan prosess yang terjadi dalam grouping data menggungkan method <b>groupby()</b> dari pandas.

jawab:
1. Pilih Kolom untuk Grouping
Deskripsi: Anda memilih satu atau beberapa kolom sebagai kunci untuk pengelompokan. Kolom-kolom ini akan digunakan untuk mengelompokkan data ke dalam grup-grup yang berbeda.
Contoh: Jika Anda memiliki DataFrame dengan kolom A dan B, dan Anda ingin mengelompokkan data berdasarkan kolom A, maka Anda akan menggunakan groupby('A').
2. Buat Objek GroupBy
Deskripsi: Setelah menentukan kolom atau kolom-kolom untuk grouping, Pandas membuat objek GroupBy. Objek ini mengorganisasi data berdasarkan nilai unik di kolom yang dipilih.
Contoh: df.groupby('A') menghasilkan objek GroupBy yang menyimpan informasi tentang grup-grup berdasarkan nilai-nilai unik di kolom A.
3. Kelompokkan Data
Deskripsi: Data di dalam DataFrame dipecah menjadi subset atau grup berdasarkan nilai-nilai unik di kolom grouping. Setiap grup berisi baris-baris dari DataFrame yang memiliki nilai yang sama di kolom tersebut.
Contoh: Jika kolom A berisi nilai 1, 2, dan 3, maka objek GroupBy akan memiliki tiga grup, satu untuk setiap nilai tersebut, masing-masing berisi baris-baris yang memiliki nilai yang sama di kolom A.
4. Terapkan Fungsi Agregasi
Deskripsi: Setelah data dikelompokkan, Anda dapat menerapkan fungsi agregasi atau transformasi ke setiap grup. Fungsi-fungsi ini dapat berupa fungsi statistik seperti mean(), sum(), count(), median(), atau fungsi kustom.
Contoh: df.groupby('A').mean() akan menghitung rata-rata untuk setiap kolom numerik di dalam setiap grup yang dibentuk berdasarkan nilai di kolom A.
5. Terapkan Fungsi Transformasi (Opsional)
Deskripsi: Selain agregasi, Anda juga bisa menerapkan fungsi transformasi yang mengubah data dalam setiap grup. Fungsi ini biasanya mengembalikan DataFrame atau Series dengan ukuran yang sama dengan data asli.
Contoh: df.groupby('A').transform(lambda x: x - x.mean()) akan mengurangi rata-rata dari setiap nilai dalam grup.
6. Hasilkan DataFrame atau Series Baru
Deskripsi: Setelah menerapkan fungsi agregasi atau transformasi, Pandas menghasilkan DataFrame atau Series baru yang berisi hasil dari operasi tersebut. Hasil ini memiliki struktur yang tergantung pada fungsi yang diterapkan.
Contoh: df.groupby('A').sum() menghasilkan DataFrame baru di mana setiap baris mewakili grup dan setiap kolom berisi hasil agregasi (jumlah) dari kolom-kolom di dalam grup tersebut.

<h2>Soal 6: Grouping All Column</h2>
    
Group data diamond berdasarkan cut (kualitas), lakukan aggregasi dengan fungsi rata2.

<h2>Soal 7: Grouping Some Column base on Multiple Criteria</h2>

<h2>Soal 8: Group By Multiple Aggregate Base on Multiple Criteria</h2>
    
Lakukan groupby pada data berdasarkan kriteria cut(kualitas) dan color (warna) secara berurutan kemudian aggregasi objek group by nya dengan fungsi rata2 dan median.

```python
#code here
relevant_df_2 = df[['cut', 'color', 'carat', 'table', 'price', 'x', 'y', 'z']]

grouped_df = relevant_df_2.groupby(['cut', 'color']).agg(
    carat_mean=('carat', 'mean'),
    carat_median=('carat', 'median'),
    table_mean=('table', 'mean'),
    table_median=('table', 'median'),
    price_mean=('price', 'mean'),
    price_median=('price', 'median'),
    x_mean=('x', 'mean'),
    x_median=('x', 'median'),
    y_mean=('y', 'mean'),
    y_median=('y', 'median'),
    z_mean=('z', 'mean'),
    z_median=('z', 'median')
)

print(grouped_df)
```

<h2>Soal 9: Group By With Different Aggregation for specific Column</h2>
    
Lakukan groupby pada data berdasarkan criteria cut(kualitas), kemudian aggreagate dengan fungsi max pada column price dan fungsi min pada column carat.

```python
#code here
grouped_df = df.groupby('cut').agg(
    price_max=('price', 'max'),
    carat_min=('carat', 'min')
)

print(grouped_df)
```

<h2>Soal 10: Group By With Custom Function</h2>
    
Lakukan groupby pada data berdasarkan criteria cut(kualitas), kemudian lakukan aggregasi dengan fungsi standar deviasi pada column carat dan fungsi rentang nilai antara max dan min pada column price.

```python
#code here
def range_func(series):
    return series.max() - series.min()

grouped_df = df.groupby('cut').agg(
    price=('price', range_func),
    carat=('carat', 'std')
)

print(grouped_df)
```
