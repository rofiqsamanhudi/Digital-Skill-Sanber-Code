<h2>Soal 3: DataFrame vs Series</h2>

- Apa perbedaan Series dengan list dan dictionary
- Apa perbedaan dataframe dan series
- Buatlah suatu data series

Jawab: 
- Apa perbedaan Series dengan list dan dictionary
  - Series:
    - Definisi: Series adalah struktur data satu dimensi dari pustaka pandas yang mirip dengan array atau list, tetapi dengan label yang dapat diatur (index).
    - Indeks: Setiap elemen dalam Series memiliki indeks yang memudahkan akses dan manipulasi data.
    - Fungsi: Memiliki banyak metode dan fungsi bawaan untuk analisis data seperti agregasi dan statistik.

  - List:
    - Definisi: list adalah struktur data bawaan di Python yang merupakan urutan elemen yang dapat diubah (mutable) dan dapat berisi berbagai tipe data.
    - Indeks: Menggunakan indeks numerik yang dimulai dari 0.
    - Fungsi: Metode terbatas dibandingkan dengan Series, seperti append(), remove(), dan sort().

  - Dictionary:
    - Definisi: dictionary adalah struktur data bawaan di Python yang menyimpan pasangan kunci-nilai (key-value).
    - Indeks: Menggunakan kunci (key) yang unik untuk mengakses nilai (value).
    - Fungsi: Metode seperti get(), items(), dan keys() untuk manipulasi data.

- Apa perbedaan dataframe dan series
  - Series:
    - Definisi: Series adalah struktur data satu dimensi yang menyimpan data dengan label indeks.
    - Dimensi: Hanya satu dimensi (seperti kolom dalam tabel).
    - Penggunaan: Biasanya digunakan untuk menyimpan satu kolom data.

  - DataFrame:
    - Definisi: DataFrame adalah struktur data dua dimensi dari pustaka pandas yang mirip dengan tabel database atau lembar kerja Excel.
    - Dimensi: Dua dimensi (baris dan kolom).
    - Penggunaan: Digunakan untuk menyimpan dan menganalisis data tabular yang lebih kompleks dengan banyak kolom dan baris.

```python
# code here
nilai_mata_pelajaran = pd.Series([85, 90, 78, 92], index=['Matematika', 'Fisika', 'Kimia', 'Biologi'])
print("Nilai Mata Pelajaran:")
print(nilai_mata_pelajaran)

cuaca_harian = pd.Series({
    'Senin': 'Cerah',
    'Selasa': 'Hujan',
    'Rabu': 'Cerah',
    'Kamis': 'Berawan',
    'Jumat': 'Cerah'
})
print("\nCuaca Harian:")
print(cuaca_harian)

rak_buku = pd.Series([15, 22, 30, 8], index=['Rak A', 'Rak B', 'Rak C', 'Rak D'])
print("\nJumlah Buku di Setiap Rak:")
print(rak_buku)
```

<h2>Soal 5: Deskriptif Statistik</h2>

Jelaskan apa itu deskriptis statistik!

Jawab: 
Statistik Deskriptif adalah cabang dari statistik yang berfokus pada pengumpulan, penyajian, dan pengorganisasian data sehingga dapat disajikan dengan cara yang mudah dipahami. Tujuan utamanya adalah untuk merangkum dan menjelaskan fitur utama dari suatu dataset dengan cara yang ringkas dan informatif.

- Contoh Penerapan: Misalkan kita memiliki dataset yang berisi nilai ujian siswa. Statistik deskriptif dapat digunakan untuk:
    - Menentukan rata-rata nilai ujian siswa (mean).
    - Mengukur seberapa bervariasinya nilai ujian siswa (standar deviasi).
    - Mengidentifikasi nilai tengah dari dataset (median).
    - Melihat apakah ada nilai yang sering muncul (modus).

<h2>Soal 6: Practice Deskriptif Statistik</h2>

- Gunakan suatu metode dari pandas untuk mengetahui deskriptif statistik dari suatu data
- Berapa nilai rata-rata dari column price
- Berapa nilai standar deviasi dari column depth
- Berapa nilai maximum dari column carat

```python
# code here
# Menampilkan deskriptif statistik dari data
print(data.describe())

# Nilai rata-rata dari column price
mean_price = data['price'].mean()
print(f'Nilai rata-rata dari kolom price: {mean_price}')

# Nilai standar deviasi dari column depth
std_depth = data['depth'].std()
print(f'Nilai standar deviasi dari kolom depth: {std_depth}')

# Nilai maksimum dari column carat
max_carat = data['carat'].max()
print(f'Nilai maksimum dari kolom carat: {max_carat}')
```

Jawab:
- Nilai rata-rata dari kolom price: 3932.799721913237
- Nilai standar deviasi dari kolom depth: 1.432621318833661
- Nilai maksimum dari kolom carat: 5.01
