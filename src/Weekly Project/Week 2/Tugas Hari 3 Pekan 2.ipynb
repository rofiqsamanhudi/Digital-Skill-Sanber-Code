<h2>Soal 2: Data Restructuring</h2>

Jelaskan itu pivoting dan melting data

- Pivoting adalah proses mengubah data dari bentuk panjang (long format) ke bentuk lebar (wide format). Ini berarti bahwa nilai-nilai dari satu kolom menjadi nama kolom baru, dan nilai-nilai terkait dipindahkan ke sel-sel yang sesuai. Fungsi utama yang digunakan dalam pandas untuk pivoting adalah pivot() atau pivot_table().

- Melting adalah proses kebalikan dari pivoting, yaitu mengubah data dari bentuk lebar (wide format) ke bentuk panjang (long format). Ini berarti bahwa kolom-kolom yang tersebar dikombinasikan menjadi dua kolom: satu untuk nama kolom asli dan satu lagi untuk nilai terkait. Fungsi utama yang digunakan dalam pandas untuk melting adalah melt().

<h2>Soal 3: Reshaping Data 1</h2>

```python
import pandas as pd
import random

data = {
    'Tipe Ujian' : ['UTS1', 'UAS1', 'UTS2', 'UAS2']*3,
    'Mata Pelajaran' : ['Fisika']*4 + ['Matematika']*4+['Statistika']*4,
    'Rata2 Waktu Belajar (menit)' : [random.choice(range(200, 800)) for i in range(12)],
    'Rata2 Nilai Ujian' : [random.choice(range(40, 100)) for i in range(12)] 
}

df = pd.DataFrame(data)
df
```

Perhatikan data di atas, kemudian ubahlah bentuknya agar sesuai yang di harapkan.

```python
#code here
pivot_df = df.pivot(index='Tipe Ujian', columns='Mata Pelajaran', values=['Rata2 Nilai Ujian', 'Rata2 Waktu Belajar (menit)'])

print(pivot_df)
```

<h2>Soal 4: Reshaping Data 2</h2>

```python
import pandas as pd

data = {
    'Hari' : ['Sabtu', 'Minggu'],
    'Kebun Binatang' : [271, 399],
    'Dufan' : [501, 700],
    'Disney' : [1000, 1001],
    'Bali' : [900, 803]
}
df = pd.DataFrame(data).set_index('Hari')
df
```

diatas adalah data tentang jumlah pengunjung di berbagai tempat hiburan, ubahlah struktur data tersebut sehingga sesuai dengan hasil yang diharapkan.

```python
#code here
df_melting = df.reset_index().melt(id_vars='Hari', var_name='Tempat Hiburan', value_name='Jumlah Pengunjung')

print(df_melting)
```

<h2>Soal 5: Data Enrichment</h2>

Lakukan binning data menggunakan qcut sehingga sesuai hasil yang diharapkan

```python
#code here
data = {
    'Tipe Ujian': ['UTS1', 'UAS1', 'UTS2', 'UAS2']*3,
    'Mata Pelajaran': ['Fisika']*4 + ['Matematika']*4 + ['Statistika']*4,
    'Rata2 Waktu Belajar (menit)': [random.choice(range(200, 800)) for _ in range(12)],
    'Rata2 Nilai Ujian': [random.choice(range(40, 100)) for _ in range(12)]
}

df1 = pd.DataFrame(data)

df1['Belajar Bin'] = pd.qcut(df1['Rata2 Waktu Belajar (menit)'], q=4, labels=False)
df1['Nilai Bin'] = pd.qcut(df1['Rata2 Nilai Ujian'], q=4, labels=False)

def kelulusan(nilai):
    if nilai >= 85:
        return 'lulus'
    elif 70 <= nilai < 85:
        return 'remedial'
    else:
        return 'tidak lulus'

df1['Keterangan kelulusan'] = df1['Rata2 Nilai Ujian'].apply(kelulusan)

df1.drop(columns=['Belajar Bin', 'Nilai Bin'], inplace=True)

print(df1)
```
